```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
#the whole solar system
Dsun1000 = open('Dsun_1000.txt','r') #here data points are taken after 10^6 s.
Dsun_data1000 = []
t = []
l = []
vals = []
for l in Dsun1000.readlines():
    vals = l.split()
    for i in range(len(vals)):
        vals[i] = float(vals[i])
    Dsun_data1000.append(vals)
Dsun1000.close()
Dsun_data1000 = np.array(Dsun_data1000)
t = np.array(Dsun_data1000[:,8])

#storing the distances from the sun for S-J-E-V-M-S
Sun = np.array(Dsun_data1000[:,0])
Jup = np.array(Dsun_data1000[:,1])
Ear = np.array(Dsun_data1000[:,2])
Ven = np.array(Dsun_data1000[:,3])
Mar = np.array(Dsun_data1000[:,4])
Sat = np.array(Dsun_data1000[:,5])
Ura = np.array(Dsun_data1000[:,6])
Nep = np.array(Dsun_data1000[:,7])


#plotting dsun against time
fig,Ax = plt.subplots(figsize=(6,7))
Ax.plot(t,Sun,label='Sun',linewidth=1.5)
Ax.plot(t,Jup,label='Jupiter',linewidth=1.5)
Ax.plot(t,Ear,label='Earth',linewidth=1.5)
Ax.plot(t,Ven,label='Venus',linewidth=1.5)
Ax.plot(t,Mar,label='Mars',linewidth=1.5)
Ax.plot(t,Sat,label='Saturn',linewidth=1.5)
Ax.plot(t,Ura,label='Uranus',linewidth=1.5)
Ax.plot(t,Nep,label='Neptune',linewidth=1.5)


Ax.legend()
#Ax.legend(loc='right',borderpad=0.3)
 
Ax.set_xlabel('time (yrs)')
Ax.set_ylabel('Distance from the Sun (AU)')

fig.tight_layout()
#plt.savefig('dsun_time1000_dt500')
plt.show()

```

```python
pcorr = open('pcorr.txt','r')
pos = []
time = [] #time array
ln = [] #lienarray to store the lines
values = []
for ln in pcorr.readlines(): #change each value to a float
    values = ln.split()
    for i in range(len(values)):
        values[i] = float(values[i])
    pos.append(values)
pcorr.close()
pos = np.array(pos)
time = np.array(pos[:,0])

#storing the positions of the Sun,Jupiter,Earth
sun = np.array(pos[:,1:3])
jup = np.array(pos[:,3:5])
ear = np.array(pos[:,5:7])

#x component of the distance from the sun 
sx = sun[:,0]
jx = jup[:,0]
ex = ear[:,0]
#y component of the distance from the sun
sy = sun[:,1]
jy = jup[:,1]
ey = ear[:,1]

fig,ax = plt.subplots()
ax.plot(sx,sy,label='Sun')
ax.plot(jx,jy,label='Jupiter')
ax.plot(ex,ey,label='Earth')
ax.set_xlabel('x component of distance(AU)')
ax.set_ylabel('y component of distance(AU)')
ax.legend()

plt.show()
```
