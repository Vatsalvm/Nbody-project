```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
Posdata = open('positions1.txt','r')
position = []
t = [] #time array
l = [] #lienarray to store the lines
for l in Posdata.readlines(): #change each value to a float
    val = l.split()
    for i in range(len(val)):
        val[i] = float(val[i])
    position.append(val)
Posdata.close()
position = np.array(position)
t = np.array(position[:,0])#the simulation runs for 30 yrs

#storing the positions of the Sun,Jupiter,Earth
Spos = np.array(position[:,1:3])
Jpos = np.array(position[:,3:5])
Epos = np.array(position[:,5:7])


#x component of the distance from the sun
sx = Spos[:,0]/(1.5e11)
jx = Jpos[:,0]/(1.5e11)
ex = Epos[:,0]/(1.5e11)
#y component of the distance from the sun
sy = Spos[:,1]/(1.5e11)
jy = Jpos[:,1]/(1.5e11)
ey = Epos[:,1]/(1.5e11)

fig,ax = plt.subplots()
ax.plot(sx,sy,label='Sun')
ax.plot(jx,jy,label='Jupiter')
ax.plot(ex,ey,label='Earth')
ax.set_xlabel('x component of distance(AU)')
ax.set_ylabel('y component of distance(AU)')
ax.legend()


plt.show()
    
```
