```python
Posdata = open('positions1.txt','r')
position = []
t = []
l = []
for l in Posdata.readlines(): #change each value to a float
    val = l.split()
    for i in range(len(val)):
        val[i] = float(val[i])
    position.append(val)
Posdata.close()
position = np.array(position)
t = np.array(position[:,0])

Spos = np.array(position[:,1:3])
Jpos = np.array(position[:,3:5])
Epos = np.array(position[:,5:7])

jx = Jpos[:,1]/(1.5e11)
sx = Spos[:,1]/(1.5e11)
ex = Epos[:,1]/(1.5e11)
fig,ax = plt.subplots()
ax.plot(t,jx,label='Jupiter')
ax.plot(t,sx,label='Sun')
ax.plot(t,ex,label='Earth')
ax.set_xlabel('Time(yr)')
ax.set_ylabel('Distance from the Sun (AU)')
ax.legend()
plt.show()
    
```
